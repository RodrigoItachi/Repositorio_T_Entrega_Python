```python
#Exercício 5.25 – Construa um programa em python que lei as 4 notas de um aluno e
#diga se ele foi aprovado ou não. Considere que para ser aprovado, necessita que a
#média das notas seja igual ou maior que 8.
```

```python
aluno = input("infomre o nome do aluno: ")
nota1 = float(input("informe a primeira nota: "))
nota2 = float(input("informe a segunda nota: "))
nota3 = float(input("informe a terceira nota:" ))
nota4 = float(input("informe a quarta nota: "))
media = (nota1+nota2+nota3+nota4/4)
if media >= 8:
 print("aluno aprovado :) :")
else:
  print("aluno reprovado: ")
```

```python
#Exercício 5.26 – Construa um programa em python que diga se um carro é
#econômico ou não. Para isso, tenha como entrada os parâmetros distância
#percorrida e quantidade, em litros, de combustível consumido. Para o carro ser
#econômico, considere que o consumo (distância percorrida / quantidade de
#combustível consumido) deve ser maior ou igual a 18.

dp = float(input("informe a distancia percorrida pelo veiculo: "))
cc = float(input("inforome a quantidade em litros consumidos: "))
consumo_c = dp/cc
print(consumo_c)
if consumo_c >= 18:
  print("este carro é  econômico")
else:
  print("carro bebedor de gasolina, compre outro")
```
